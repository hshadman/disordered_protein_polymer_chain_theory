# formula from 2012 macromolecules elder jayaraman paper
$$
asphericity=(R_1)-\frac{1}{2}(R_2+R_3) \\  
acylindricity=(R_2-R_3) \\
RSA=\frac{asphericity^2 + 0.75acylindricity^2}{R_g^4}
